## 10장 노트정리 ##

##### 컴퓨터 네트워크란 
- 정보와 자원 공유를 위해 링크로 연결된 노드들의 집합

###### 구성요소 #####
1. 메세지
- 메세지는 패킷으로 구성
- payload: 문자,숫자,소리,영상,그림 또는 멀티미디어 데이터등을 전송하는데 필요한 해더 등을 제외한 순수한 데이터를 의미
2. 송신자
- 메세지를 보내는 장치
3. 수신자
- 메세지를 받는 장치
4. 전송매체
- 송신자에서 수신자까지 메세지를 전달하는 물리적인 경로
- 꼬임선, 동축선 광케이블, 레이저 또는 무전전파
5. 프로토콜
- 데이터 통신을 위한 수행 규칙들의 집합(상호규약)
- 구문: 전송되는 데이터의 구조 및 형식
- 의미: 전송되는 데이터의 의미
- 순서: 데이터의 전송 순서




##### 프로토콜의 구조
* 계층 구조
- 복잡한 시스템을 계층별로 나누어 모듈화
- 모듈 사이의 인터페이만 유짛하면 유지관리, 수정용이
- n 계층의 역할: n계층은 (n-1)계층의 서비스를 받아 (n-1)계층에게 서비스를 제공
- n 계층 프로토콜: n 계층끼리 사용하는 통신 규칙
- 인터페이스: 상하위 계층 사이의 접속 방법

- c/s모델: 요청과 응답/ 서버에 자원 집중
- p2p모델: 동등한 요청/응답, 자원분산

##### OSI 7계층 모델
- 7 응용계층: WWW,FTP,Telnet과 같은 응용 프로그램
- 6 표현계층: 데이터의 암축과 암호화 기능 수행
- 5 세션계층: 세션을 관리하고 대화 단위로 데이터 송수신을 수행
- 4 전송계층: 종간단 데이터 전송
- 3 네트워크계층: 데이터 전송 경로 설정
- 2 데이터링크계층: 데이터 오류 검출 및 흐름 제어
- 1 물리계층: 데이터를 전기적 신호로 변환


1. 물리계층
- 실제 장치를 연결하여 데이터를 전송하기 위해 필요한 기능과 처리 절차 규정
- 물리계층의 주요 기능
1) 통신 장치와 전송매체간의 인터페이스 특성을 규정
2) 비트의 표현: 비트를 전송하기 위해 전기적 똔느 광학적인 신호로 부호화
3) 데이터 속도: 신호가 유지되는 비트의 주기를 규정
4) 비트의 동기화: 송신자와 수신자 사이에 비트 주기의 동기화
