<br> 객체 지향 프로그래밍 <br>
* 객체지향 프로그램은 객체의 속성(변수)과 동작(함수)을 나타내는 클래스를 정의하고, 이로부터 객체를 생성하여 작성한 프로그램
* 클래스와 객체는 풀빵틀,풀빵
* 클래스의 동작을 나타내는 함수를 메소드라 한다
* 클래스는 다른 클래스로부터 상속받을 수 있다. 상속 클래스는 부모 클래스가 갖지 않는 속성과 메소드를 가질 수 있다
* 객체는 클래스의 속성과 메소드를 갖는다
* 객체는 일반 변수처럼 사용된다. 함수의 인자로 사용되고 함수에서 객체를 반환할 수 있다 튜플이나 딕셔너리의 요소로 사용될 수 있다

#### 클래스의 정의
    class 클래스 이름(부모클래스):
    클래스 몸체

* 클래스는 class 키워드로 정의된다. 부모 클래스는 없으면 생략
* 클래스 몸체는 변수와 메소드로 구성

#### 클래스 객체
* 클래스를 인스턴스화하여 만들어진것을 객체라한다
* 객체는 클래스의 변수와 메소드에 접근할 수 있다

#### 클래스 예
class Car :
    def __init__(self, color, speed): 
        self.color = color 
        self.speed = speed
    def speedUp(self, v): 
        self.speed = self.speed + v
        return self.speed
    def speedDown(self, v):
        self.speed = self.speed – v 
        return self.speed
* self.variable로 정의된 변수는 클래스 전체에서 유효하다. 클래스 속성은 인스턴스 변수로 정의한다
self.가 없는 변수는 메소드 내에서만 유효한 지역변수다

#### 클래스 초기화 함수
### 초기화 함수(__init__())
* 객체를 생성할때 자동 실행 메소드
* 인스턴스 변수를 초기화할때 사용
* self를 첫번째 매개변수로 가지며 self는 클래스 객체를 나타낸다
* 인스턴스 변수는 클래스 내부의 모든 메소드에서 사용할 수 있다

#### 클래스 속성
* 클래스 변수와 인스턴스 변수

#### 클래스 상속
* 상속은 기존에 만들어진 클래스로부터 속성과 메소드를 이어받고 자신이 필요한 기능을 추가하는 기법이다
* 상위 클래스 = 부모 클래스 또는 super class
* 하위 클래스 = 자식 클래스 또는 sub class

#### 메소드 오버라이딩
* 자식 클래스에서 부모클래스의 메소드를 수정해 다시 정의하는것

#### 다형성
* 전달된 인자에 따라 함수 또는 연산의 기능이 달라지는 기능

#### 가시성
* public: 이름 앞에 밑줄 없음 ->어디서나 접근 가능
* protected: 이름 앞에 밑줄 1개 ->하위 클래스에서 접근 권장, 외부에서 직접 접근은 권장하지 않음
* private: 이름 앞에 밑줄 2개 ->클래스 내부에서만 접근 가능,이름이 맹글링 되어 외부/하위 클래스에서 직접 접근 어렵게 만듬

#### 추상 함수
* @abstractmethod 데코레이터를 사용해 정의
* 추상 메서드는 반드시 하위 클래스에서 구현해야 하며, 직접 인스턴스화 할 수 없음

#### 정적 멤버
* static, member는 클래스 전체에서 공유되는 변수
* 클래스 변수, 클래스 메소드라고 함
* 메서드의 경우 @staticmethod 데코레이터를 사용해 정적 메서드를 정의
* 정적 변수는 클래스 안에 self가 아닌 일반 변수로 선언하고 클래스.변수 형태로 접근

#### 데코레이터
* 함수나 클래스 동작을 수정하거나 확장할때 사용하는 문법
* 함수 정의 위에 @데코레이터 이름 형태 사용
* 데코레이터 함수 정의는 다음 형식을 따라 wrapper 이름의 멤버 메소드를 정의

#### 기존 함수를 확장방법
* 1. 함수 재할당
* 작식이 된 함수명을 기존 함수명으로 그대로 사용하고 싶을때-> 기존 클라이언트 코드 변경 없이 원본 함수명
* 2. 다른 함수명
* 장식이 된 함수명을 새로운 함수명으로 명명하여 사용해도 좋을때-> 새롭게 클라이언트 함수를 짤때
